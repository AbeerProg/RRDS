```python
!pip install spacy
!pip install nltk
!python -m spacy download en_core_web_sm

import pandas as pd
import spacy
import nltk
from collections import Counter
from nltk.tokenize import word_tokenize

# Load SpaCy English model
nlp = spacy.load('en_core_web_sm')

# Download necessary NLTK data
nltk.download('punkt')

# Load the dataset
df = pd.read_excel('1.xlsx')

# Display the first few rows of the dataset
print(df.head())


def calculate_repeated_word_count(text):
    # Tokenize the text into words
    words = word_tokenize(text.lower())  # Convert to lower case to normalize

    # Count the words
    word_counts = Counter(words)

    # Count words that are repeated (appear more than once)
    repeated_word_count = sum(1 for word, count in word_counts.items() if count > 1)

    return repeated_word_count

# Apply the repeated words calculation to the preprocessed text
df['repeated_word_count'] = df['text'].apply(calculate_repeated_word_count)

# Display the DataFrame with the new columns
print(df[['text', 'repeated_word_count']])

# Save the updated DataFrame to a new Excel file
df.to_excel('repatedWordscount.xlsx', index=False)
```
