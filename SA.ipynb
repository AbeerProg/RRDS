```python
import pandas as pd

df = pd.read_excel('Evaluation.xlsx')
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Sentiment Distribution by Label
sns.histplot(data=df, x='S', hue='LABEL', kde=True)
plt.title('Sentiment Score Distribution by Label')
plt.show()

```

```python

# Example: Length Distribution by Label
sns.boxplot(x='LABEL', y='WC', data=df)
plt.title('Review Length Distribution by Label')
plt.show()

```

```python
# Example: Capital Words Ratio by Label
sns.histplot(data=df, x='RWC', hue='LABEL', kde=True)
plt.title('Repated Words Ratio Distribution by Label')
plt.show()
```

```python
# Example: Capital Words Ratio by Label
sns.histplot(data=df, x='WC', hue='LABEL', kde=True)
plt.title(' Review Length Distribution by Label')
plt.show()
```

```python
# Check for duplicate reviews with different labels
duplicate_reviews = df[df.duplicated(subset=['text'], keep=False) & df['LABEL'].nunique() > 1]
print(duplicate_reviews)

```

```python
# Check for outliers in sentiment score or other features
sns.boxplot(x='LABEL', y='S', data=df)
plt.title('Sentiment Score Boxplot by Label')
plt.show()
```
